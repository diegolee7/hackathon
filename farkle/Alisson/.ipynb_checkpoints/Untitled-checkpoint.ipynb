```python
import pandas as pd
import numpy as np
import json
import re
#from unidecode import unidecode
```

```python
js = json.load(open('../hackathon/opinioes.json', encoding="utf8"))
df = pd.read_csv('../hackathon/notas.csv',sep=';')
```

```python
op = list()
lista = ['']*16
for opiniao in js:
    lista[0] = opiniao[0]
    lista[1] = opiniao[1]
    lista[2] = opiniao[2].rsplit(' ', 1)[0]
    lista[3] = opiniao[2].rsplit(None, 1)[-1]
    divisor = opiniao[3].split(' - ')
    lista[4] = divisor[0]
    lista[5] = divisor[1].rsplit(' ', 1)[0]
    lista[6] = divisor[1].rsplit(None, 1)[-1]
    if 'kmCarro anterior:' in opiniao[4]:
        divisor = opiniao[4].split(' - ')
        lista[7] = re.findall('.+([0-9]+).+',divisor[0])[0]
        divisor = divisor[1].split(' kmCarro anterior: ')
        lista[8] = divisor[0]
        lista[9] = divisor[1]
    else:
        lista[7] = opiniao[4]
        lista[8] = ''
        lista[9] = ''
    lista[10] = opiniao[5].replace('Prós:','')
    lista[11] = opiniao[6].replace('Contras:','')
    lista[12] = opiniao[7].replace('Defeitos apresentados:','')
    lista[13] = opiniao[8].replace('Opinião Geral:','')
    if ' ' in opiniao[9]:
        divisor = opiniao[9].split(' ')
        lista[14] = divisor[0]
        lista[15] = divisor[1]
    else:
        lista[14] = opiniao[9]
        lista[15] = ''
    op.append(lista.copy())
opinioes = pd.DataFrame(op,columns=['id','comentario','carro','ano','proprietario','cidade','estado','posse','quilometragem','anterior','pros','contras','defeitos','comentario','data','hora'])
opinioes
#opinioes.to_csv('opinioes.csv',index = False)
```
