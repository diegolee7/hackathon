```python
import pandas as pd
import numpy as np
import json
#from unidecode import unidecode
```

```python
dados = pd.read_csv('../dados.csv',encoding="ISO-8859-1")
```

```python
dfCarro = dados.groupby(['carro']).mean()
dfCarro.describe()
```

```python
dfCarro['opiniao geral polaridade'] = ((dfCarro['opiniao geral polaridade'] - dfCarro['opiniao geral polaridade'].min()) / (dfCarro['opiniao geral polaridade'].max() - dfCarro['opiniao geral polaridade'].min()))*10
dfCarro['opiniao geral subjetividade'] = ((dfCarro['opiniao geral subjetividade'] - dfCarro['opiniao geral subjetividade'].min()) / (dfCarro['opiniao geral subjetividade'].max() - dfCarro['opiniao geral subjetividade'].min()))*10
```
