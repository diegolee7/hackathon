```python
# encoding: utf-8
# encoding: iso-8859-1
# encoding: win-1252
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv = pd.read_csv('../hackathon/notas.csv', delimiter=';')

cols = ['ID','Estilo', 'Acabamento', 'PosDir', 'Instrumentos', 'Interior', 'Malas', 'Desempenho', 'Motor', 'Cambio', 'Freios', 'Suspensao', 'Consumo', 'Estabilidade', 'CustoBeneficio', 'Recomendacao']


dfNotas = pd.DataFrame(csv)
userAtt = list()
user = 0
        
csv['nota']

notas = list()
users = list()

for i in range(len(csv)):
    if i%15 == 0:
        notas.append(csv['ID'][i])
        notas.append(csv['nota'][i])
    else:
        notas.append(csv['nota'][i])
    
    if i%15 == 14:
        users.append(notas.copy())
        del (notas[:])
        
dfUsers = pd.DataFrame(users, columns=cols)
dfUsers['AvGeral'] = dfUsers.mean(axis=1)
dfUsers.head()
```

```python
# Correlação Quesitos/Avaliação Geral
for i in list(dfUsers)[1:-1]:
    print (i, dfUsers[i].corr(dfUsers['AvGeral']))
```
