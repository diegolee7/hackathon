```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../dados.csv',encoding = "ISO-8859-1")
data.sort_values(['estado'])
```

```python
dfCarro = data.groupby('carro').mean()
dfCarro
```

```python
dfCarro['comentario polaridade'] = 10*(dfCarro['comentario polaridade'] - dfCarro['comentario polaridade'].min())/(dfCarro['comentario polaridade'].max()-dfCarro['comentario polaridade'].min())
dfCarro['comentario subjetividade'] = 10*(dfCarro['comentario subjetividade'] - dfCarro['comentario subjetividade'].min())/(dfCarro['comentario polaridade'].max()-dfCarro['comentario subjetividade'].min())

dfCarro
```
